```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ptitprince
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/plthiyagu/Personnel/master/Dataset/Iris.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
pd.plotting.andrews_curves(df, 'Species', colormap='YlOrRd')
```

```python
plt.figure(figsize = (12,8))
ptitprince.RainCloud(data = df, x = 'Species', y = 'SepalLengthCm', orient = 'h')
plt.show()
```
