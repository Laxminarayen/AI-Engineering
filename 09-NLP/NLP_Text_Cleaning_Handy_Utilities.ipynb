```
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import fetch_20newsgroups
newsgroups_train = fetch_20newsgroups(subset="train")
```

```
type(newsgroups_train)
```

```
import pandas as pd
df= pd.DataFrame([newsgroups_train.data,newsgroups_train.target.tolist()])
```

```
df
```

```
df= pd.DataFrame([newsgroups_train.data,newsgroups_train.target.tolist()]).T
df.columns=['text','target']

targets= pd.DataFrame(newsgroups_train.target_names)
targets.columns=['title']

ngout=pd.merge(df,targets,left_on='target',right_index=True)
```

```
ngout['title'].value_counts()
```

```
ngout_final= ngout.query('title in["rec.sport.hockey","comp.graphics","talk.politics.misc"]')
```

```
pd.set_option('display.max_colwidth',-1)
ngout_final
```

```
import re
def clean_text(in_string):
  remove_email= re.sub('\S*@\S*\s?','',in_string)
  remove_nl=re.sub('\s+',' ',remove_email)
  remove_other= re.sub("\'|\>|\:|\.","",remove_nl)
  return remove_other
```

```
ngout_final.iloc[10]['text']
```

```
clean_text(ngout_final.iloc[10]['text'])
```

```
import texthero as hero
```

```
ngout_final['tfidf']=(
    ngout_final['text']
    .pipe(hero.clean)
    .pipe(hero.tfidf)
)
```

```
ngout_final['kmeans']=(
    ngout_final['tfidf']
    .pipe(hero.kmeans,n_clusters=3)
    .astype(str)
)
```

```
ngout_final
```

```
ngout_final['pca']=ngout_final['tfidf'].pipe(hero.pca)
```

```
hero.scatterplot(ngout_final,'pca',color='kmeans',title='Newsgroup clusters')
```
