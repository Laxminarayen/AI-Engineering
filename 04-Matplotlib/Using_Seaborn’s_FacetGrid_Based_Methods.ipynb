```python
import seaborn as sns
df = sns.load_dataset('mpg')
# df.head()
sns.catplot(data=df, x="cylinders", y="displacement", col = 'origin', hue = 'model_year');
```

```python
import seaborn as sns
df = sns.load_dataset('mpg')
# df.head()
sns.catplot(data=df, x="cylinders", y="displacement", col = 'origin', hue = 'model_year', kind = 'box');
```
