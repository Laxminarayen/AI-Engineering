```
import pandas as pd
import sweetviz
```

```
trainfilename='https://raw.githubusercontent.com/plthiyagu/Personnel/master/Dataset/House%20Prices%20train.csv'
testfilename='https://raw.githubusercontent.com/plthiyagu/Personnel/master/Dataset/House%20Prices%20test.csv'
```

```
train= pd.read_csv(trainfilename)
test= pd.read_csv(testfilename)
train.tail()
```

```
my_report = sweetviz.analyze([train, "Train"],target_feat='SalePrice')
```

```
my_report1 = sweetviz.compare([train, "Train"], [test, "Test"], "SalePrice")
```

```
my_report1.show_html('Report.html')
```
