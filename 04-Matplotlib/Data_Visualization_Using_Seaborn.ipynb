```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename='https://raw.githubusercontent.com/plthiyagu/Personnel/master/Dataset/ToyotaCorolla.csv'
cars_data= pd.read_csv(filename)
```

```python
cars_data.head()
```

Scatter plots can be used to show a linear relationship between two or three data points using the seaborn library.

 A Scatter plot of price vs age with default arguments will be like this:



```python
plt.style.use("ggplot")
plt.figure(figsize=(8,6))
sns.regplot(x = cars_data["Age"], y = cars_data["Price"])
plt.show()
```

```python
sns.lmplot(x='Age', y='Price', data=cars_data,
          fit_reg=False,
          hue='FuelType',
          legend=True,
          palette="Set1",height=6);
```

```python
plt.figure(figsize=(8,6))
sns.distplot(cars_data['Age'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.distplot(cars_data['Age'],kde=False)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.distplot(cars_data['Age'],kde=False,bins=5)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="FuelType", data=cars_data)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="FuelType", data=cars_data,
             hue="Automatic")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=cars_data["Price"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=cars_data["FuelType"], 
           y=cars_data["Price"],
           )
plt.show()  
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="FuelType", 
            y="Price",
            data=cars_data,
            hue="Automatic"
           )
plt.show()
```

```python
f, (ax_box,ax_hist) = plt.subplots(2, gridspec_kw={"height_ratios":(.15,.85)})
```

```python
f, (ax_box,ax_hist) = plt.subplots(2, gridspec_kw=
                                  {"height_ratios":(.15,.85)})
sns.boxplot(cars_data["Price"], ax=ax_box)
sns.distplot(cars_data["Price"], ax=ax_hist, kde=False)
plt.show()
```

```python
sns.pairplot(cars_data, kind="scatter", hue="FuelType")
plt.show()
```
